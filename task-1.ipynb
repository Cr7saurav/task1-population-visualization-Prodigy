```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (skip first 4 metadata rows in World Bank format)
file_path = "C:\\Users\\saura\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)

# Select year 2020 data
df_2020 = df[['Country Name', '2020']].dropna()

# ------------------------------
# 1. Histogram - Distribution of populations in 2020
# ------------------------------
plt.figure(figsize=(10, 6))
plt.hist(df_2020['2020'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Country Populations in 2020', fontsize=14)
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ------------------------------
# 2. Bar Chart - Top 10 most populous countries in 2020
# ------------------------------
top10_2020 = df_2020.sort_values(by='2020', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(top10_2020['Country Name'], top10_2020['2020'], color='orange', edgecolor='black')
plt.title('Top 10 Most Populous Countries in 2020', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
